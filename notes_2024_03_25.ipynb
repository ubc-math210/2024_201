```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la
```

## Interpolation

```python
x = np.array([-2.,-1.,0.,1.,2.])
y = np.array([1.,0.,2.,-1.,2.])
xs = np.linspace(-2,2,100)

# Interpolate with trig functions
A = np.column_stack([np.ones(len(x)),np.cos(x),np.sin(x),np.cos(2*x),np.sin(2*x)])
c = la.solve(A,y)
ys = c[0] + c[1]*np.cos(xs) + c[2]*np.sin(xs) + c[3]*np.cos(2*xs) + c[4]*np.sin(2*xs)
plt.plot(xs,ys)

# Interpolate with polynomial
A = np.vander(x,increasing=True)
c = la.solve(A,y)
ys = c[0] + c[1]*xs + c[2]*xs**2 + c[3]*xs**3 + c[4]*xs**4
plt.plot(xs,ys)

plt.plot(x,y,'.')
plt.show()
```

## Least Squares

```python
x = np.array([0.,1.,2.,3.,4.,5])
y = np.array([5.,4.,4.2,3.7,2.9,3.2])

A = np.column_stack([np.ones(len(x)),x])
c = la.solve(A.T@A,A.T@y)
xs = np.linspace(-1,6,10)
ys = c[0] + c[1]*xs

plt.plot(xs,ys)
plt.plot(x,y,'.')
plt.show()
```
