## Plotting Procedure

To plot a function $y = f(x)$ we follow the steps:

1. Create the vector of $x$ values in the interval of the plot using `np.linspace`.
2. Create the vector of $y$ values using vectorized arithmetic operations and functions.
3. Plot with `plt.plot(x,y)`.
4. Add style, legend, title, labels, etc.

## First Order ODE

Plot solutions of $y' = -y^2 t$ for initial conditions $y(0) = 0.1,0.2,\dots,2.0$.

\begin{align*}
y' &= -y^2 t \\
-\frac{y'}{y^2} &= t \\
- \int \frac{dy}{y^2} &= \int t dt \\
\frac{1}{y} &= \frac{t^2}{2} + C \\
y &= \frac{1}{t^2/2 + C}
\end{align*}

And note that $y(0) = 1/C$.

```python
plt.figure(figsize=(5,3))
t = np.linspace(0,10,100)
for y0 in np.linspace(0.1,2,20):
    C = 1/y0
    y = 1/(t**2/2 + C)
    plt.plot(t,y,'b')
plt.plot(t,t*0,'b') # Plot the solution y(t) = 0
plt.grid(True)
plt.show()
```

## Parametric Curves

A parametric curve in 2D is given by $(x(t),y(t))$. It's like a trajectory or orbit. The simplest parametric curve is the circle:

\begin{align*}
x(t) &= \cos(t) \\
y(t) &= \sin(t) \\
\end{align*}

```python
t = np.linspace(0,2*np.pi,50)
x = np.cos(t)
y = np.sin(t)
plt.figure(figsize=(4,4))
plt.plot(x,y), plt.axis('equal'), plt.grid(True), plt.xlim([-2,2]), plt.ylim([-2,2])
plt.show()
```

## Hypotrochoid

```python
t = np.linspace(0,10*np.pi,1000)
R = 7; r = -4;
d = 5
x = (R - r)*np.cos(t) + d*np.cos((R - r)*t/r)
y = (R - r)*np.sin(t) - d*np.sin((R - r)*t/r)
plt.plot(x,y); plt.axis('equal');
plt.show()
```
