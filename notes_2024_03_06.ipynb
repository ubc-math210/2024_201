## Slope Fields

The slope field of a first order differential equation is a graphical method to visualize solutions. Create a grid of points in the $ty$-plane and the plot a line of slope $f(t,y)$ (and length $L$) at each point $(t,y)$ in the grid. The result shows the shape of solutions of the differential equation $y' = f(t,y)$.

For example, plot the slope field of $y' = -y^2 t$ and plot the exact solution for the unique solution satisfying $y(0)=1$.

```python
f = lambda t,y: -y**2*t

t = np.linspace(0,4,21)
y = np.linspace(-2,2,21)
L = 0.5*(t[1] - t[0])
for i in range(len(t)):
    for j in range(len(y)):
        m = f(t[i],y[j])
        theta = np.arctan(m)
        dt = L*np.cos(theta)
        dy = L*np.sin(theta)
        plt.plot([t[i],t[i]+dt],[y[j],y[j]+dy],'b')
y = lambda t: 1/(t**2/2 + 1)
plt.plot(t,y(t),'r')
plt.show()
```

## Euler's Method

Euler's method essential constructs an approximation by following the slopes in the slope field. More precisely, Euler's method applied to $y' = f(t,y)$, $y(t_0)=y_0$, is the recursive sequence:

$$
y_{n+1} = y_n + f(t_n,y_n)h \ \ , \ \ \ t_n = t_0 + nh \ \ , \ \ \ y_0 = y(0)
$$

Each value gives an approximation of the value of the soluiton $y_n \approx y(t_n)$.

Apply Euler's method to $y' = -y^2t$, $y(0)=1$ for $N=100$ steps from $t_0=0$ to $t_f=2$. Plot with the exact solution.

```python
f = lambda t,y: -y**2*t
t0 = 0
y0 = 1

N = 100
tf = 2
h = (tf - t0)/N
t = np.linspace(t0,tf,N+1)
y = np.zeros(N+1)
y[0] = y0

for n in range(N):
    y[n+1] = y[n] + f(t[n],y[n])*h

plt.plot(t,y,label="Euler's Method")
T = np.linspace(t0,tf,200)
Y = 1/(T**2/2 + 1)
plt.plot(T,Y,'r',label='Exact Solution')
plt.legend()
plt.show()
```
