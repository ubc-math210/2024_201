```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as spi
```

## SciPy Solver

The function `scipy.integrate.odeint` numerically approximates solutions of differential equations very efficiently and accurately.

```python
# y' = f(y,t) , y(t0) = y0
f = lambda y,t: -y
y0 = 1; t0 = 0; tf = 2; N = 3;
t = np.linspace(t0,tf,N+1)

y = spi.odeint(f,y0,t)
plt.plot(t,y,'.',label='odeint')

T = np.linspace(t0,tf,100)
plt.plot(T,np.exp(-T),label='Exact')
plt.legend()
plt.show()
```

```python
np.abs(np.exp(-tf) - y[-1])
```

```python
# y' = f(y,t) , y(t0) = y0
f = lambda y,t: np.cos(y) + np.sin(t*y)
y0 = 1; t0 = 0; tf = 20; N = 200;
t = np.linspace(t0,tf,N+1)

y = spi.odeint(f,y0,t)
plt.plot(t,y)
plt.show()
```

```python
a = 1; b = 1; c = 1;
F = lambda t: np.cos(t)
f = lambda u,t: np.array([u[1],(F(t) - b*u[1] - c*u[0])/a ])
u0 = [0,0]
t0 = 0; tf = 20; N = 200;
t = np.linspace(t0,tf,N+1)
u = spi.odeint(f,u0,t)
plt.plot(t,u)
plt.show()
```
