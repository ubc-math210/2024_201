```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0
b = 1
N = 13
dx = (b - a)/N
x = np.linspace(a,b,N+1)
f = lambda x: 1/(1 + x**5)
y = f(x)
TN = np.sum(y[1:] + y[:-1])/2*dx
print(TN)
```

```python
x = np.linspace(0,1,100)
y = -10*x**3*(2 - 3*x**5)/(1 + x**5)**3
K2 = np.max(np.abs(y))
print(K2)
plt.plot(x,y)
plt.show()
```
